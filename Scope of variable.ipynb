### Global : Outside the function
### Local : Inside the function
### Non-local : Inside the Nested Function

##### --We cannot directly update the global variables as local or Non-Local But we can only use the same. we need to use the keyword global

```python
x = 10
def outer(a):
    c = a+10
    def inner(b):
        k = b+10
        print('This is from inner Function:',k)
    inner(c)
    print('This is from outer function:',c)
outer(x)
print(x)
```

```python
x = 10
def outer(a):
    c = a+x
    def inner(b):
        k = c+x
        print('This is from inner Function:',k)
    inner(c)
    print('This is from outer function:',c)
outer(x)
print(x)
```

```python
# Global Variables trying to update in local scope but it raises UnboundlocalError

x = 10
def outer(a):
    x = a+x   #cannot access local variable 'x' where it is not associated with a value
    c = x+2
    def inner(b):
        c = c+x
        print('This is from inner Function:',c)
    inner(c)
    print('This is from outer function:',c)
outer(x)
print(x)
```

```python
# Global Variables trying to update in local scope but it raises UnboundlocalError

x = 10
def outer(a):
    global x
    x = a+x   
    c = x+2
    def inner(b):
        nonlocal c
        c = c+b
        print('This is from inner Function:',c)
    inner(c)
    print('This is from outer function:',c)
outer(x)
print(x)
```

### Pass by Value and Pass by Reference
  - **Pass By Value** : In pass by Value ,the argument objects are share the object(argument variables) copy value to function definition parameter.
       - All Immutable Data types(i.e String,Tuple,int) behvae like pass by value
       - After Function Execution, the original value not effects

  - **Pass By Reference** : In pass by Reference, The argument objects shaers the object(argument variable) refernce to functio  definition parameter. So after the function execution original values will effects.

      - All Mutable datatypes(i.e list,set,dictionary) behave like pass by reference.


```python
# Pass by value--> Shallow Copy
# Immutables Can't update the Values so it is pass by values --> Immutable = String,Tuple,Primitive Data types

s = 'Codegnan'
def replace(string, old, new):
    res = string.replace(old,new)
    return res
print(replace(s, 'n', 'Z'))
print(s)    
```

```python
# Pass by Refernce--> Deep copy
lst = [1,2,3,4]
def add_ele(arr):
    arr.append(100)
    return arr
print(add_ele(lst))
print(lst)
```

```python
def alphabet_pattern(n):
    if n* n>26:
        return
    ch = 65
    for i in range(n):
        for j in range(n):
            print(chr(ch),end=' ')
            ch +=1
        print()
alphabet_pattern(5)
```

```python
# filter(fun,seq)

def even(x):
    if x % 2==0:
        return x
lst = [1,2,3,4,5,6]
print(filter(even,lst))
```

```python
# filter(fun,seq)

def even(x):
    if x % 2==0:
        return x
lst = [1,2,3,4,5,6]
print(list(filter(even,lst)))
```

```python
# map(fun,seq)

def add(x,y):
    return x+y
print(list(map(add,[1,2,3,4],[1,2,3])))
```
