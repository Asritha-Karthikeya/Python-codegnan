```python
#1.Find length of number
#2.Find sum of digits
#3.Find Reverse of a number
#4.Check weather the number is palindrome
#5.Check weather the number is armstrong or not
```

```python
#1.Find length of number
num=int(input())
count=0
while num>0:
    digit=num%10
    num//=10
    count+=1
print(count)
```

```python
#1.Find length of number in function
def length_of_num(num):
    count=0
    while num>0:
        digit=num%10
        num//=10
        count+=1
    return count
    
num=int(input())
print(length_of_num(num))
```

```python
#2.Find sum of digits
num=int(input())
total=0
while num>0:
    digit=num%10
    total=total+digit
    num//=10
print(total)
```

```python
#2.Find sum of digits in function
def sum_digits(num):
    total=0
    while num>0:
        digit=num%10
        total=total+digit
        num//=10
    return total
    
num=int(input())
print(sum_digits(num))
```

```python
#3.Find Reverse of a number
num=int(input())
rev=0
while num>0:
    digit=num%10
    rev=rev*10+digit
    num//=10
print(rev)
```

```python
#3.Find Reverse of a number in function
def reverse_num(num):
    rev=0
    while num>0:
        digit=num%10
        rev=rev*10+digit
        num//=10
    return rev
num=int(input())
print(reverse_num(num))
```

```python
#4.Check weather the number is palindrome
def reverse_num(num):
    original=num
    rev=0
    while num>0:
        digit=num%10
        rev=rev*10+digit
        num//=10
    if original==rev:
        return "Palindrome"
    else:
        return "Not Palindrome"
num=int(input())
print(reverse_num(num))
```

```python
#5.Check weather the number is armstrong or not
def armstrong(num):
    original=num
    sum_val=0
    power=len(str(num))
    while num>0:
        digit=num%10
        sum_val+=digit**power
        num//=10
    if original==sum_val:
        return "Armstrong"
    else:
        return "Not Armstrong"
num=int(input())
print(armstrong(num))
```
