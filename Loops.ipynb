#### Loop : Loop is a repeated task , It performs operations on different values.
  - Types of Loops:
     - for Loop
     - while loop


    - For Loop:- In for loop Iterations will be only auto Increment / auto Decrement

        - Syntax:
           - for (keyword)  iterator (variable name)  in (membership operator)  iterable (range()/set()/,list[]/tuple()/string()dict{}):
           -      # block of statements
           -   else:(optional)
           -      # Block of code

    - range function syntax :

         - range(start,stop,step)

```python
# Syntax for for loop

- For Loop:-
        - Syntax:
           - for (keyword)  iterator (variable name)  in (membership operator)  iterable (range()/set()/list[]/tuple()/string()/dict{}):
                 # block of statements
              else:(optional)
                 # Block of code
```

```python
#Syntax for range function

- range function syntax :
         - range(start,stop,step)

    start (optional)
    step- (optional)
```

```python
# N NaturalNumbers
n = int(input())
for num in range(1,n+1):
    print(num,end=' ')
print('Done')
```

```python
# List of Even and Odd 
n = int(input())
n2 = int(input())
even=[]
odd=[]
for num in range(n,n2+1):
    if num %2:
        odd.append(num)        
    else:
        even.append(num)   
print('Even',even)
print('odd',odd)
```

```python
#Find the factors of a number
n = int(input())
result = []
for num in range(1,n+1):
    if n % num ==0:
        result.append(num)
print(result)
```

```python
n = int(input())
result = []
for num in range(1,n//2+1):
    if n % num ==0:
        result.append(num)
result.append(n)
print(result)
```

```python
#Sum of n natural numbers
n = int(input())
res = 0
for num in range(1,n+1):
    res = res + num
print(res)
```

```python
#Sum of n natural numbers
n = int(input())
res = 0
for num in range(1,n+1):
    res = res + num
    print(f'{res} + {num} = {res}')
print(res)
```
