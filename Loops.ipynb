     - for Loop

           - for (keyword)  iterator (variable name)  in (membership operator)  iterable (range()/set()/,list[]/tuple()/string()dict{}):

## Codeforces Problems 

```python
# A) N Natural Numbers
n = int(input())
for num in range(1,n+1):
    print(num)
```

```python
# B) Check how many even numbers before given number if not found print -1
N = int(input())
found = False
for num in range(2,N+1):
    if num % 2 ==0:
        print(num)
        found = True
if not found:
    print(-1)
```

```python
# C) Check count of given numbers weather How mny even,odd,positive,negative
N = int(input())
n = list(map(int,input().split()))
even = odd = negative = positive = 0
for num in n:
    if num %2:
        odd +=1
    else:
        even +=1
    if num  > 0:
        positive +=1
    elif num < 0:
        negative +=1
print(f'Even: {even}')
print(f'Odd: {odd}')
print(f'Positive: {positive}')
print(f'Negative: {negative}')
         
```

```python
# E) Gretest number among given numbers
N = int(input())
n = list(map(int,input().split()))
res = n[0]
for num in n:
    if num > res :
        res = num
print(res)
```

```python
# O) Pyramid
n = int(input())
for x in range(1,n+1):
    print('*'*x)
    
```

```python
# T) Another Shape
n = int(input())
for x in range(1,n+1):
    print(' ' *(n-x) + '*' *(2*x - 1))
   
```

```python
n = int(input())
for x in range(n , 0,-1):
    print(' ' *(n-x) + '*' *(2*x - 1))
```

```python
# W) Shape 3
n = int(input())
for x in range(1,n+1):
    print(' ' *(n-x) + '*' *(2*x - 1))
for x in range(n , 0,-1):
    print(' ' *(n-x) + '*' *(2*x - 1))
```
