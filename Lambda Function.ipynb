#### Anonymous Function: Anonymous Function doesn't have function name and this function are implemented with 'lambda' keyword.

#### Syntax : 
             - lambda arguments : Expression
   -lambda : keyword ; arguments : variable name 

```python
# Anonymous Function
# Find Square of a function

square = lambda x : x*x
print(type(square))
print(square(2))
```

### Single line if-else statement
 - Syntax :
            -  True_exp if condition else False_exp

```python
# Find weather a number is even or odd

even_or_odd = lambda x :'even' if x%2==0 else 'odd'
print(even_or_odd(12))
print(even_or_odd(1))
```

```python
# Check the Square root of a number, if number is even otherwise square of a number

num = lambda x : x**0.5 if x%2==0 else x**2
print(num(36))
print(num(7))
```

```python
# Check weather the number is perfect square or not

num = lambda x : 'Perfect sqaure' if x**0.5 == int(x**0.5) else 'Not perfect sqaure'
print(num(16))
print(num(3))
print(num(39))
print(num(36))
```

```python
# Filter the even numbers from the list using lambda and filter function

lst = [1,2,3,4,5,6]
print(list(filter(lambda x: x%2==0,lst)))
```

```python
# Find the square of each number in list using lambda function and map function

lst = [2,4,5,6,3]
print(list(map(lambda x:x*x, lst)))
```

```python
lst = [2,7,5,6,3]
print(list(map(lambda x:'even' if x%2 ==0 else 'odd' , lst)))
```

## Recursion :
  - Function calls itself either directly or indirectly
#### Types of Recursion :
  - Direct Recursion
  - Indirect Recursion

 #### In Recursion , mainly it have two blocks,
  - Base case
  - Recursive call case.

### 1) Direct Function : Function calls itself directly

```python
# Direct funtion : Function calls itself

def fact(n):
    if n ==0 or n ==1:
        return 1
    return n * fact(n-1) # --> 5 * fact(4)--> 20 *fact(3) -->
x = int(input())
fact(x)
```

```python
# Return 1 to n Numbers using recursive function

def num(n):
    if n==1:  # base case
        print(n)
        return 1
    print(n,end=' ')
    num(n-1)
x=int(input())
num(x)
```

```python
# return n to 1 Numbers using recursive function

def num(n):
    if n==1:
        print(n,end=' ')
        return 
    num(n-1)
    print(n,end=' ')
x= int(input())    
num(x)
```

```python
# Return 1 to n Numbers using recursive function in a list

lst =[]
def num(n):
    if n==1:  # base case
        return [1]
    lst = num(n-1)
    lst.append(n)
    return lst
    
x=int(input())
num(x)
```

```python
# Return 1 to n Numbers using recursive function in a list

lst =[]
def num(n):
    if n==1:  # base case
        return [1]
    return [n] + num(n-1)
    
x=int(input())
num(x)
```
