#### Set : Set is Unordered Collection of Different Datatype unique elements. It is mutable and defines with '{}'.
  - Characteristics:
    - Set is unordered Collection
    - Set is Mutable
    - It allows only unique elements
    - Set not allows list as element, Bcoz list is unhasable datatype.
    - Set does not supports indexing and slicing
    - All set elements can be used as dictionry keys.

   - In The sets the values will be written using the hash values. If You you remove an element from the set , the positions of another elements will change and again if you add the same removed value it again replace the position at which it got removed. 
        

#### Set Operations
  - Union
  - Intersection
  - Difference
  - Symmetric 
  - Sub-set
  - Super-set
  - Disjoint set

```python
#Set is unordered Collection
#It returns only unique elements

s = {10,3,4,5,'hi',5,6,'k'}
print(s, type(s))
```

#### Set Methods 
  - add(val) : It adds the value to the set of the value is already not exists in the set
  - update (new_set) : It adds all new set elements
  - remove(val) : It removes the specified value from the set if it is present in set otherwise ''' KeyError '''
  - discard(val) : It Removes the value if it is present in set, otherwise nothing
  - union(new_set) : It returns the all unique elements from the sets.
  - Intersection(new_set) : It returns the only common elements from all sets
  - Difference(new_set) : It returns the set elements those are not present in new_set
  - symmetric _difference(new_set) : it returns the common elements from the sets

```python
s = {10,3,4,5,'hi',5,6,'k'}
# add(val)
s.add(7) ## {'k', 3, 4, 5, 6, 7, 10, 'hi'}
print(s)

s.update({}) ## {'k', 3, 4, 5, 6, 7, 10, 'hi'}
print(s)

# Update(new_set)
s.update({8,'bye',2}) ## {'k', 2, 3, 4, 5, 6, 7, 8, 10, 'bye', 'hi'}
print(s)


#remove(val)
s.remove('bye')
s.remove(5)
print(s)  ## {'k', 2, 3, 4, 6, 7, 8, 10, 'hi'}

# Discard(val)
s.discard(2)
print(s)

s.discard(2)
print(s)

s.remove(5)
print(s)  # KeyError: 5 . Because we already removed the 5 from the set . if again we tried to remove the same element it raises the key error 
```

```python
s = {'k', 3, 4, 6, 7, 8, 10, 'hi'}
s.discard(3,4)
print(s)
```

```python
#
a = {1,2,3}
b = {4,3,2,6}

print(a.union(b))
print(a|b) # Union in symbol

print(a.intersection(b))
print(a & b)

print(a.difference(b))
print(a-b)

print(a.symmetric_difference(b))
print(a ^ b)

```

```python
a = {1,2,3}
b = {4,3,2,6}

print(a.issubset({1,2}))
print(a.issubset({1,2,3,4}))

print({1,2,3,4}.issuperset(a))

print({1}.isdisjoint(b))
```
