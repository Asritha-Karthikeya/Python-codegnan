```python
# Sum of digits
def sum_of_digits(n):
    s = 0
    while n>0:
        digit = n%10
        s = s+digit
        n = n//10
    return s
num = int(input())
print(sum_of_digits(num))
```

```python
# product of digits
def pro_of_digits(n):
    s = 1
    while n>0:
        digit = n%10
        s = s*digit
        n = n//10
    return s
num = int(input())
print(pro_of_digits(num))
```

```python
# Reverse of a number
def reverse(n):
    rev = 0
    while n>0:
        digit = n%10
        rev = rev * 10+digit
        n = n//10
    return rev
n = int(input())
print(reverse(n))
```

```python
# Reverse of a number even if it is negavtive
def reverse(n):
    rev = 0
    if n<0:
        sign = -1
        n = n*sign
    else:
        sign = 1
        n = n* sign
    while n>0:
        digit = n%10
        rev = rev * 10+digit
        n = n//10
    return rev*sign
n = int(input())
print(reverse(n))
```

```python
# Chcek weather a given number is palindrome or not
def palindrome(n):
    rev = 0
    temp = n
    while n>0:
        digit = n%10
        rev = rev * 10+digit
        n = n//10
    return 'Palindrome' if temp == rev else 'Not-palindrome'
num = int(input())
print(palindrome(num))
    
```

```python
def armstrong(n):
    s = 0
    power = len(str(n))
    temp = n
    while n>0:
        digit = n%10
        s = s+digit**power
        n = n//10
    return 'Armstrong' if s == temp else ' Not-armstrong'
num = int(input())
print(armstrong(num))
```

```python
# Sum of digits of a given number using recursion
def sum_of_digits(n,res=0):
    if n ==0:
        return res
    return sum_of_digits(n//10,res+n%10)
n=int(input())
print(sum_of_digits(n))
```

```python
# Sum of digits of a given number using recursion
def sum_of_digits(n,res=1):
    if n ==0:
        return res
    return sum_of_digits(n//10,res*n)
n=int(input())
print(sum_of_digits(n))
```

```python
# Length of a given number
def len_of_digits(n,res=0):
    if n ==0:
        return res
    return len_of_digits(n//10,res+1)
n=int(input())
print(len_of_digits(n))
```

```python
# reverse of a given number
def rev_of_digits(n,res=0):
    if n ==0:
        return res
    return rev_of_digits(n//10,res*10+n%10)
n=int(input())
print(rev_of_digits(n))
```

```python
def fibonacci(n):
    a, b = 0, 1
    for i in range(n):
        print(a, end=" ")
        a, b = b, a + b

fibonacci(6)
```

```python
def fib(n):
    if n <= 1:
        return n
    return fib(n-1) + fib(n-2)

for i in range(6):
    print(fib(i), end=" ")
```
