#### Functions :- Function is reusable block of statement and it is used to perform specific task.
  - Characteristics:
      - Reusable Block Code
      - Reduce Program Code
      - Modularity and easy to debug
      - Increase Program readability and understability
      - Function definition starts with 'def' Keyword

#### Syntax:
   - #function def
   - def function_name(parameters):
       - #block code
       - return values/Expessions (optional)
   - #function calling
   - function_name(arguments)


#### Parameters: It is nothing but inputs to the function definition those where represented inside the function definition ().
#### and it is optional

#### Arguments: The Variables are Passed to function calling inputs, those variables calls as arguments. and it is optional

#### Types of Functions :
  - User defined functions.
  - Built-in functions.

#### User defined functions: The function definition are implemented by user or programmer.
  - The user defined functions are classified in to 4 types based on parameters and return values.
      - without return and without parameters.
      - without return and with parameters
      - with return and without parameters
      - with return and with parameters

#### No.of Parameters and Arguments should be same

```python
# find count of special character in a string
def spl_char_count(n):
    count = 0
    for char in n:
        if not char.isalnum() and not char.isspace():
            count+=1
    return count
string = input()
res = spl_char_count(string)
print(res)
```

## Types of Arguments
  - Positional arguments 
  - Keyword Arguments
  - Default Arguments
  - Variable Length Positional Arguments
  - Variable Length Keyword Arguments

#### Positional arguments :- 
  The Values are assign to parameters exactly specified oreder.The first argument value assign to first parameters, second argument to second parameter,and so on.

```python
# Types of Arguments
# 1) Positional Arguments

# function def
def sub(x,y):
    z = x-y
    return z
    
#function calling
a, b = 10, 30
res1 = sub(a,b)
res = sub(b,a)
print(res)
print(res1)
```

```python
# 2) Keyword Arguments

# Function def
def sub(x,y):
    z = x-y
    return z
    
#function calling
a, b = 10, 30
res1 = sub(x=a,y=b)
res = sub(y=b,x=a)
print(res)
print(res1)
```

```python
def add(x,y,z):
    return x+y+z
def add4(x,y,z,s):
    return x+y+z+s
# Function calling
a =int(input())
b = int(input())
c = int(input())
d = int(input())
res = add(a,b,c)
res1 = add4(a,b,c,d)
print(res)
print(res1)
```

```python
# 3) Default Arguments
def add(x,y, z = 0,s = 0):
    return x+y+z+s
    
print(add(10,35))
print(add(2,6,3))
print(add(23,5,7,4))
```

```python
def mul(x=1,y=1, z = 1,s =1):
    return x*y*z*s
    
print(mul(10,35))
print(mul(2,6,3))
print(mul(23,5,7,4))
print(mul(5))
print(mul(2,5,3,1))
```

```python
# 4) Variable length Positional Arguments
def add(x,y=0, *args):
    #print(args,type(args))
    s = 0
    s +=x
    s +=y
    for num in args:
        s+=num
    return s

print(add(2))
print(add(2,4))
print(add(20,10,2,4,2,3,3))
```

```python
# 4) Variable length Positional Arguments
def add(x,y=0, *args):
    #print(args,type(args))
    s = 0
    s =x+y
    for num in args:
        s+=num
    return s

print(add(2))
print(add(2,4))
print(add(20,10,2,4,2,3,3))
```

```python
# 5) Variable length keyword arguments
def details(name,**kwargs):
    info ={}
    info['Name'] =name
    info['Class']= kwargs.get('cls','no class')
    if 'cls' in kwargs:
        kwargs.pop('cls')
    marks = kwargs.values()
    info['avg'] = 0
    if marks:
        avg = sum(marks)/ len(marks)
        info['avg'] = avg
    return info
    
print(details('Asritha'))
print(details('Asritha',cls = 10))
print(details('Asritha',m1=90,m2=74,m3=92))
print(details('Asritha', cls = 10,m1=90,m2=74,m3=92,m4=80))
```
