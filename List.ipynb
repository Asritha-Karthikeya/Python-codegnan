#### List methods
- Adding elements to list
    - append(ele) : It adds the element at the end of the list. Can only insert single value
    - extand(list_else) : It adds all list elements at end of the list. Can add List of elements
    - insert(pos, val) : It inserts the value at particular position
- removing elements from the list
    - remove(val) : It removes the first occurance value if value is present  in list, otherwise ValueError will raise
    - pop() : It removes and returns the last element in the list
    - pop(pos) : It removes and returns specified position element from the  list.
- index finding for an element
    - index(val) : It return the first occurance value index position
- Sorting list
    - sort() : It sorts the original list into asscending order
    - sort(reverse= True ) : It sorts the original list into descending order

```python
# Shallow Copy
l1 = [1,2,3]
l2 = [1,2,3]

print(l1==l2)
print(l1 is l2)
print(id(l1) , id(l2))

l2 = l1 # Shallow Copy will effects the another list bcoz of the assigning same reference
print(l1 is l2)

l1.append(6)
print(l2)
```

```python
# Deep Copy
l1 = [1,2,3]
l2 = l1[:] # Deep copy will never effects the another list , Difference references
print(l1 == l2)
print(l1 is l2)

print(id(l1) , id(l2)) # Same values but different locations/ references

l1.append(6) 
print(l2) 
```
