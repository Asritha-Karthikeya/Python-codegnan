#### Dictionary : Dictionary is Collection of Key-Value Pairs. It is Mutable and Defines with '{}'.
 - EX : { Key : Value} ---> One Item
   
    - Characteristics:
        - Dictionary is ordered collection 
        - Dictionary is Mutable
        - In Dictionary Keys are Unique
        - List Can't be Used as Dictionary key because Unhashable data types
        - Dictionary Value allows duplicate and Lists also
        - Dictionary values can access by using Key.
        - Dictionary Doesn't Support Indexing.

```python
# Dictionry keys are Unique
details = { 'name':'Asritha' , 'Course' : 'Python', 'Age':22, 'name': 'Radha'} 
# If No.of Keys Mentioned with same key name then it returns only last updated value

print(details)
```

```python
# Dictionary is Mutable
# Updating Key-values
# If Dictionary contains key , it updates with value , otherwise it create new item to the Dictionary

details = { 'name':'Asritha' , 'Course' : 'Python', 'Age':22}
details['Age'] = 29
details['Location'] = 'Andhra Pradesh' # Item Assignment
print(details)

print(type(details))
```

#### Dictionary Methods 
    - Keys() : It returns list dictionaary keys
    - values() : It returns list dictionary values
    - items() : It returns list of tuples key-value pairs
    - get(key, default_val) : It returns the values of key if it is present in dictionary , otherwise default value
    - update(new_dictionary) : Dictionary updates with new_dictionary items
    - pop(key) : It returns the value of key and removes the item from dictionary 
    - popitem() : It returns and remove the last item from dictionary

```python
details = { 'name':'Asritha' , 'Course' : 'Python', 'Age':22, 'name': 'Radha','Location': 'Andhra Pradesh'} 
keys = details.keys()
values = details.values()
items = details.items()
print(keys, type(keys))
print(values, type(values))
print(items, type(items))
print(list(keys))
```

```python
# get(key,default_val)
details = { 'name':'Asritha' , 'Course' : 'Python', 'Age':22, 'name': 'Radha'}
print(details.get('name'))
print(details.get('location', 'India'))
print(details.get('location')) # None
print(details.get('Age',25))
#details.pop('name')
print(details)
```

```python
#update()

details = { 'name':'Asritha' , 'Course' : 'Python', 'Age':22 }
det = {'2nd name': 'Karthikeya', 'Batch' : 15} 
details.update(det)
print(details)

#details.pop('name')
print(details , details.pop('Batch'))

print(details.popitem()) # Returns Tuple of Key- Value pair
```
