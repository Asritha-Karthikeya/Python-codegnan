#### String methods
- Case conversion methods
    - Upper() : It returns the upper converted string
    - Lower() : It returns the lower case
    - title() : It converts the first character in ech word into upper cse nd remaining all are into lower case
    - swapcase(): It converts the lowercse charcter into  uppercase nd viceversa
    - captilize() : It converts first character of string into uppercase and remaining all are into lower case 

```python
s = 'The sky is blue and Ocean is BLUE'
print(s.upper())
print(s.lower())
print(s.title())
print(s.swapcase())
print(s.capitalize())
```

```python
s = 'Codegnan batch @ 51'
print(s.upper())
print(s.lower())
print(s.title())
print(s.swapcase())
print(s.capitalize())
```

```python
# Substring index finding
# find(substring): It return the first occurance index position of substring , otherwise -1
# index(substring): It return the first occurance index position of substring , otherwise value error
s = 'codegnan'
print(s[0])
print(s.index('n'))
print(s.find('n'))
#print(s.find(0)) #TypeError 
print(s.find('gn'))
print(s.find('A')) # As we know in slicing and finding the error won't occur
print(s.find('3'))
print(s.index('A'))
```

- *** White space or pattern removing from string ***
    - strip(chars) : Returns a copy of the string with leading and trailing whitespace removed.
    - lstrip() : Returns a copy of the string with leading (left side of the sentence) whitespace removed.
    - rstrip() : Returns a copy of the string with trialing (right side of the sentence) whitespace removed.

```python
s ='     Codegnan    '
print(s)
print(s.strip())
print(s.lstrip())
print(s.rstrip())
```

```python
s ='     Code   gnan    '
print(s)
print(s.strip())
print(s.lstrip())
print(s.rstrip())
```

```python
s ='@@@@@    Code@@gnan   @@@'
print(s)
print(s.strip('@'))
print(s.lstrip('@'))
print(s.rstrip('@'))
```

#### String Manipulation methods
- replace(old,new) : It replaces old string with the new string
- split() : It splits the string into list of substrings
- "sep".join(list) : It combines the all list of words into list of words with separator

```python
s= 'The sky is blue and ocean is blue'
print(s.replace('is' ,'ISs'))
print("ABCBCBC".replace('CBC', '@'))
print(s.split())
print("% &G hf ey 7 ng".split())
print(s.index('blue'))
print(s.split('blue'))
```

```python
list_words = ['The', 'sky', 'is', 'blue', 'and', 'ocean', 'is', 'blue']
print("".join(list_words))
print(" ".join(list_words))
print("_".join(list_words))
```

#### String Checking Methods
- This all methods return boolean values as result
    - islower() : It returns true if entire string is lowercase
    - isupper() : It returns true if entire string is uppercase
    - istitle() : It returns true if entire string is title manner
    - isalpha() : It returns true if entire string contains only Alphabets
    - isalnum() : It returns true if entire string contains alphabets and digits
    - isdigit() : It returns true if entire string contains only digits
    - startswith(sub_string) : It returns true if the string starts with sub_string
    - endswith() : It returns true if the string ends with sub_string

```python
s = 'CodeEdu67hund'
print(s.islower())
print(s.isupper())
print(s.istitle()) # Occur false when digits encountered
print(s.isalnum()) # occur false when speacial charceters encountered
print('CODE'.isupper())
print('code'.islower())
print('fgjhu'.isalpha())
print('67938'.isdigit())
print('gdyh867t'.isdigit())
print('gdyh867t'.isalnum())
```
