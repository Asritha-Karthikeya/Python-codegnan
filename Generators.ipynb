### Generators: 
#### Generators is special type of function if the function contains ' Yield' Keyword those functions call it as  Generators.It creats a iterable object

#### Generators Take less memory compared to functions
### Generators are called as lazy iterators or operators

#### Syntax:
  - def generator_name(parameters):
      - #Block code
      - yield val/Exp
      - yield val/exp
      - #block code
  - #calling
  - generator_name(arguments)
   

```python
def seq(x):
    a = x+5
    yield a
    yield x+a
    yield x+2
res = seq(5)
print(res)
print(next(res))
```

```python
def seq(x):
    a = x+5
    yield a
    a = a+2
    yield x+a
    yield x+2
res = seq(5)
print(next(res))
print(next(res))
print(next(res))
print(next(res)) #StopIteration 
```

```python
def seq(x):
    a = x+5
    yield a
    yield x+a
    yield x+2
res = seq(5)
print(res)
for num in res:
    print(num)
```

```python
def num_1toN(n):
    while n>0:
        yield n
        n+=1
```
