#### Type Conversion / Casting : Converting one datatype value into another datatype
  - There are two types of type conversions
      - implicit type conversion : automatically converts one datatype to another datatype
          - Ex: 3 + 2.5 = 5.
          - Ex: 2/3= 0.66...
      - explicit type conversion : Here conversions happened through the built in datatype functions
      -  Built-in types = int(),str(),list(),tuple(),dict(),set()
         - Ex: ('10')#string +20 #int = int('10')+20 = 30
         - Ex : '10' + 20 = '10' + str(20) = '1020'

      

#### Type conversions will only possible on ..
  - Int - string, float, bool
  - float - int, string, bool
  - bool - int, string
  - string - list, tuple , set
  - list - string , tuple, set
  - tuple - string, list, set
  - set - string, tuple, list
  - dict - list, string

```python
a=45
b='67'
print(a + int(b))
print(str(a) + b)

c=8.9
print(a+c)

```

```python
#type conversion and type casting  @ Input formating
#Int - string, float, bool
num = int(input())
print(type(num))
val = float(num)
print(type(val),val)
string_num = str(num)
print(type(string_num), string_num)
print(bool(num))
```

```python
#float - int, string, bool
num = float(input())
print(type(num),num)
val = int(num)
print(type(val),val)
string_num = str(num)
print(type(string_num), string_num)
print(bool(num))
```
