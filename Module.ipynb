### Module : 
  - A Module is nothing but it is a python source file consists of variables, functions and classes. The file extension should be in '.py' .
  - One module source can get in some other modules by using ' import' Keyword.

#### Why modular Coding:
  - Easy to debug
  - Easy to Maintain
  - Increases Readability
  - Multiple people can works at a time.
  - Reusability.

#### Types of Modules :
  - User defined Module
  - Built-in Modules

#### User Defined Modules : The Modules are defined by the Users or Programmers.
  - Ex: small_cal.py 

#### Built-in Modules:
  - random(): To generate random numbers
  - numpy(): numerical Calculations
  - pandas(): Data manipulation
  - matplotlib(): Data Visualization
  - seaborn(): Data Visualization
  - math(): For math calculations
  - scikit_learn(): For ML algorithms
  - collections():
  - tinker(): 
  - datetime(): For date and time

```python
import random
print("Welcome to the who will reach to in less chances")
steps = 0
chances = 0
while True:
    input("To Throw a dice for that click on enter button")
    dice_val = random.randint(1,6)
    chances += 1
    if steps + dice_val <=100:
        steps += dice_val
    if steps > 90:
        print(f'hello buddy u r at steps {steps} and total chances are {chances} and you need {100-steps} to reach top')
    else:
        print(f'hello buddy u r at steps {steps} and total chances are {chances}')
    if steps == 100:
        break
```
