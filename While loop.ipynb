#### While Loop : While loop Executes the Block of statement until condition become false
  - Syntax:
     - #initialization
     - while condition:
     -    #block of code #Increment/decrement
     -  else:
     -    #block of code

#### Steps to follow :
  - Initialization
  - Condition
  - Increment/Decrement

#### Statements:
  - While Loops we will prefer when you don't know the exact iterations count
  - In while loops manually we can perform increment or decrement


```python
n = int(input())
num = 0
while num<n:
    if num %2 ==0:
        print(num ,end =' ')
        num +=2
    else:
        num+=1
        print(num , end=' ')
```

```python
# N Natural Numbers
n = int(input())
num =1
while num<n+1:
    print(num,end= ' ')
    num+=1
```

```python
# List of Even and Odd Numbers
n = int(input())
n1 = int(input())
even = []
odd = []
num =n
while num<=n1:
    if num%2:
        odd.append(num)
        num+=1
    else:
        even.append(num)
        num+=1
print('Even',even)
print('Odd',odd)
```
