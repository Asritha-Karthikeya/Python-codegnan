#### Tuple : Tuple is a ordered Collection of different data type elements. It is immutable and defines with "()".
   - Characteristics: 
        - Tuple is Ordered Collection  
        - It supports with Slicing and Indexing
        - Tuple is Immutable
        - Tuple is Heterogeneous data type # Allows Different types of data
        - Tuple takes less memory compared to List
        - Tuple operations are faster compared to list
        - Tuple allows duplicate values
          

### Tuple Supports Only 
    - Count 
    - Index

```python
# Tuple is Immutable
t = (1,4,7,4,35,'Hi',[1,235,],32)
print(t)
print(t[6])
t[6].append(9) # We cannot add elements to Tuple , But we added the value to the list but the positions are not Changed. So it is Called Immutable
print(t)

#t[0] = 10 # 'tuple' object does not support item assignment

# If Tuple contains any element as mutable elements (i.e list, set , and dictionary), you can perform operations on that ,
##   but tuple Doesnot Change remaining elements position and tuple object length also. 
```

```python
# Read Tuple of Integers
t = tuple(map(int,input().split()))
```

```python
# Creating empty Tuple
t1 = ()
t2 = tuple()
print(t1,type(t1) , type(t2))
```

```python
# Single Element representation of Tuple
t1 = (10)
print(t1 , type(t1))

t1 = (10,)
print(t1 , type(t1))
```
