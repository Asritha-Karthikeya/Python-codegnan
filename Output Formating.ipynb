#### Output Formating
  - Using Old style Formatting (Using commas)
  - Using modulo operator
      - Access Specifiers
         - integer :- %d
         - string :- %s
         - float :- %f
  - Using F- Format Method
  - Using dot Format Method.

```python
name = 'Asritha'
age = 20
print('My name is',name,'and I am', age,'years old')
```

```python
name = 'Asritha'
age = 20
print('My name is' + name + 'and I am' + age + 'years old')
```

```python
name = input()
age = int(input())
print('My name is',name,'and I am', age,'years old')
```

```python
#Using Modulo Operator
name = input()
age = int(input())
print('My name is %s and I am %d years old'%(name,age))
```

```python
#Using f-Format Method
name = input()
age = int(input())
print('My name is {} and I am {} years old'.format(name,age))
```

```python
name = input()
age = int(input())
print('My name is {} and I am {} years old'.format(age,name))
```

```python
#Using f- format Method
name = input()
age = int(input())
print(f'My name is {name} and I am {age} years old')
```

```python
name = 'Asritha'
age = 20.6478473
print(f'My name is {name} and I am {age:.2f} years old')
```
